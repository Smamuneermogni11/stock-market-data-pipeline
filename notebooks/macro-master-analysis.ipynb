```python
msft = msft.rename(columns={"MSFT_Date": "Date"})
interest_rates = interest_rates.rename(columns={"InterestRate_Date": "Date"})
inflation = inflation.rename(columns={"inflation_Date": "Date"})
vix = vix.rename(columns={"vix_Date": "Date"})
qqq = qqq.rename(columns={"qqq_Date": "Date"})
oil = oil.rename(columns={"oil_Date": "Date"})
gas = gas.rename(columns={"gas_Date": "Date"})
```

```python
datasets = {
    "msft": msft,
    "vix": vix,
    "qqq": qqq,
    "interest_rates": interest_rates,
    "inflation": inflation,
    "oil": oil,
    "gas": gas
}
```

```python
for name, df in datasets.items():
    print("\n", name)
    print(df.columns)
```

```python
inflation.head(10)
```

```python
skip_first_two = ["msft", "vix", "qqq", "oil", "gas"]

for name, df in datasets.items():
    if name in skip_first_two:
        df = df.iloc[2:].reset_index(drop=True)
        datasets[name] = df
```

```python
msft = datasets["msft"]
vix = datasets["vix"]
qqq = datasets["qqq"]
oil = datasets["oil"]
gas = datasets["gas"]
```
